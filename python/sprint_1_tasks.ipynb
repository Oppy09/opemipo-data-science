```python
# Task 1.4
# Assignment (=, +=, -=)
x = 10
x += 5   # x becomes 15
x -= 2   # x becomes 13

# Comparison (==, !=, <, >, <=, >=)
print(x == 13)  # True
print(x != 10)  # True
print(x > 10)   # True
print(x < 20)   # True
print(x >= 13)  # True
print(x <= 5)   # False

# Logical (and, or, not)
print(x > 10 and x < 20)   # True (Both are true)
print(x == 10 or x == 13)  # True (At least one is true)
print(not x == 13)         # False (Reverses the true result)

# Membership (in, not in)
allowed_numbers = [10, 11, 12, 13, 14]
print(x in allowed_numbers)      # True
print(x not in allowed_numbers)  # False
```

```python
# Task 1.7: Loops
# A for loop iterating over a list
items = ["apple", "banana", "cherry"]
for item in items:
    print(item)

# A for loop using range()
for i in range(3):
    print(i)  # Prints 0, 1, 2

# A while loop with a condition
count = 0
while count < 3:
    print(count)
    count += 1

# The use of break and continue
for num in range(5):
    if num == 2:
        continue  # Skips the rest of the loop for '2' and moves to '3'
    if num == 4:
        break     # Exits the loop completely when it hits '4'      
    print(num)
```

```python
# Task 1.8
# Function that takes a name and prints a personalized greeting
def print_greeting(name):
    print(f"Hello, {name}! Welcome to Python.")

# Function that takes two numbers and returns their sum, difference, and product
def math_operations(a, b):
    sum_val = a + b
    diff_val = a - b
    prod_val = a * b
    return sum_val, diff_val, prod_val

# A third simple function that returns the square of a number
def square_number(n):
    return n * n

print_greeting("Alice")

s, d, p = math_operations(10, 5)
print(f"10 and 5, Sum: {s}, Difference: {d}, Product: {p}")

squared = square_number(4)
print(f"The square of 4 is: {squared}")
```
