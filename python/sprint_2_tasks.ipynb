```python
# Task 2.3: Sample Project
import csv

students = []      
subjects = set() 

def add_student(name, age, course, grade):
    student_id = (len(students) + 1, "2025")  
    student = {
        "id": student_id,
        "name": name,
        "age": age,
        "course": course,
        "grade": grade
    }
    students.append(student)
    subjects.add(course)

def view_students():
    if not students:
        print("No students in the system yet.")
        return
        
    for student in students:
        id_num, year = student["id"]
        print(f"ID: {id_num} ({year}), Name: {student['name']}, Age: {student['age']}, Course: {student['course']}, Grade: {student['grade']}")

def search_student(search_name):
    found = False
    for student in students:
        if search_name.lower() in student["name"].lower():
            id_num, year = student["id"]
            print(f"Found: {student['name']} (ID: {id_num}), Course: {student['course']}, Grade: {student['grade']}")
            found = True
            
    if not found:
        print("No student found with that name.")

def update_grade(search_name, new_grade):
    for student in students:
        if student["name"].lower() == search_name.lower():
            student["grade"] = new_grade
            print(f"\nUpdated {student['name']}'s grade to {new_grade}.")
            return
            
    print("\nError: Student not found.")

def display_courses():
    if not subjects:
        print("No courses added yet.")
        return
        
    for course in subjects:
        print(f"{course}")


def save_to_file():
    with open("students.csv", "w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=["id_num", "year", "name", "age", "course", "grade"])
        writer.writeheader()
        
        for student in students:
            id_num, year = student["id"] # Unpack the tuple
            writer.writerow({
                "id_num": id_num,
                "year": year,
                "name": student["name"],
                "age": student["age"],
                "course": student["course"],
                "grade": student["grade"]
            })

def load_from_file():
    try:
        with open("students.csv", "r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                student = {
                    "id": (int(row["id_num"]), row["year"]), 
                    "name": row["name"],
                    "age": int(row["age"]),
                    "course": row["course"],
                    "grade": row["grade"]
                }
                students.append(student)
                subjects.add(row["course"])
    except FileNotFoundError:
        pass


add_student("Alice Smith", 20, "Computer Science", "A")
add_student("Bob Jones", 22, "Mathematics", "B")
add_student("Charlie Brown", 19, "Physics", "A")
add_student("Diana Prince", 23, "Computer Science", "A")
add_student("Evan Wright", 21, "Biology", "C")

view_students()
search_student("Alice")
update_grade("Bob Jones", "A")
display_courses()
save_to_file()

```
